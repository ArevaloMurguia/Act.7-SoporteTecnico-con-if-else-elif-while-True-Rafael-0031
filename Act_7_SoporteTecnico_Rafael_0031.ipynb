### SoporteTecnico con if, else, elif , while  True

```python
# UIII Act 7
# IA con if, else, elif y while True
# Tema: Soporte Técnico

print("================================")
print(" IA DE SOPORTE TECNICO ")
print("================================")

while True:

    print("\n1 Computadora lenta")
    print("2 Problema de internet")
    print("3 Virus")
    print("4 No enciende")
    print("5 Salir")

    op = input("Opción: ")

    if op == "1":

        memoria = int(input("Memoria disponible GB: "))

        if memoria < 4:
            print("La IA recomienda aumentar memoria RAM")

        else:
            print("La IA recomienda limpiar archivos")

    elif op == "2":

        internet = input("¿Hay conexión? si/no: ")

        if internet == "no":
            print("Reiniciar módem")

        else:
            print("Revisar cables")

    elif op == "3":

        antivirus = input("¿Tiene antivirus? si/no: ")

        if antivirus == "no":
            print("Instalar antivirus")

        else:
            print("Actualizar antivirus")

    elif op == "4":

        bateria = int(input("Nivel de batería: "))

        if bateria < 20:
            print("Conectar cargador")

        else:
            print("Revisar sistema")

    elif op == "5":

        print("Programa terminado")
        break

    else:
        print("Opción incorrecta")
```
